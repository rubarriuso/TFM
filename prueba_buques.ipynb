```python
import pandas as pd

```

```python
df = pd.read_csv(
    './resources/Vessel_registry/EU_fishing_vessel_registry_nov_2025.csv',
    sep=';',
    quotechar='"',
    on_bad_lines='skip',
    engine='python'
)
df["Main fishing gear"].unique()
```

```python
gear_list=['OTB', 'TBN', 'TBS', 'OTP', "OTT", "PTB", "TBB"]
secondary_gear_list =['OTB', 'TBN', 'TBS', 'OTP', "OTT", "PTB", "TBB", "NK"]
bottom_trawlers = df[df['Main fishing gear'].isin(gear_list)]
bottom_trawlers_strict =bottom_trawlers[
    bottom_trawlers['Subsidiary fishing gear 1'].isin(secondary_gear_list)| df['Subsidiary fishing gear 1'].isna() &
    bottom_trawlers['Subsidiary fishing gear 2'].isin(secondary_gear_list)| df['Subsidiary fishing gear 2'].isna() 
    ]
bottom_trawlers_strict
```

```python

bottom_trawlers_MMSI = bottom_trawlers_strict["MMSI"].dropna().unique().astype(int).astype(str).tolist() #all bottom trawlers MMSI
#Tendremos tambien que verificar si sufrio una modificacin y dej de ser un bottom trawler en algn momento
#una vez tenemos el MMSI, con el podemos filtrar ese MMSI y ver si dej de ser un bottom trawler
#Tambien se debe ver si ese buque empez a ser bottom trawler en algn momento.
#Solucion: hacer una lista para usar en cada ao, de los cuales se eliminan o se aaden los barcos cada ao
#en funcin de si se modifican. (Los datos de esfuerzo de GFW se obtienen por un ao)
```

```python
bottom_trawlers_strict = bottom_trawlers_strict.dropna(subset=['MMSI'])
bottom_trawlers_strict["MMSIstr"] = bottom_trawlers_strict["MMSI"].astype(int).astype(str).tolist()
```

```python
bottom_trawlers_strict_df = bottom_trawlers_strict[["Country of Registration", "Event", "Event Start Date", "Event End Date", "Name of vessel", "Vessel Type", "Main fishing gear", "Subsidiary fishing gear 1", "LOA", "Tonnage GT", "Power of main engine", "MMSIstr"]]
```

```python
bottom_trawlers_strict_df.to_csv('./data/strict_bottom_trawlers.csv', index=False)
```
