```python
import copernicusmarine
import xarray as xr
import numpy as np
from dask.diagnostics import ProgressBar
import xesmf as xe
```

```python
min_lon, min_lat, max_lon, max_lat = [-69.6095, -60., -50., -32.4483] 
#estas coordenadas est√°n extraidas del area de la fao 41 la 3.1 3.2 y 2.3
start_date = "2012-01-01"
end_date = None  # hasta la fecha mas reciente disponible
```

```python
# Mass concentration of chlorophyll a in sea water CHL [mg/m3]
copernicusmarine.subset(
    dataset_id="cmems_obs-oc_glo_bgc-plankton_my_l4-multi-4km_P1M",
    variables=["CHL"],
    minimum_latitude=min_lat, minimum_longitude=min_lon,
    maximum_latitude=max_lat, maximum_longitude=max_lon,
    start_datetime=start_date, end_datetime=end_date,
    file_format="zarr",
    output_filename = "cmems_obs-oc_glo_bgc-plankton_my_l4-multi-4km_P1M.zarr",
    output_directory="./resources/copernicus_marine_service",
    chunk_size_limit=50_000_000,
    overwrite=True,
    dry_run=True # False to download and True to only simulate
)
```

```python
#Primary productivity of biomass expressed as carbon PP [mg/m2/day]
copernicusmarine.subset(
    dataset_id="cmems_obs-oc_glo_bgc-pp_my_l4-multi-4km_P1M",
    variables=["PP"],
    minimum_latitude=min_lat, minimum_longitude=min_lon,
    maximum_latitude=max_lat, maximum_longitude=max_lon,
    start_datetime=start_date, end_datetime=end_date,
    file_format="zarr",
    output_filename = "cmems_obs-oc_glo_bgc-pp_my_l4-multi-4km_P1M.zarr",
    output_directory="./resources/copernicus_marine_service",
    chunk_size_limit=50_000_000,
    overwrite=True,
    dry_run=True # False to download and True to only simulate
)
```

```python
# Mass concentration of suspended matter in sea water SPM [g/m3]
# Secchi depth of sea water ZSD [m]
copernicusmarine.subset(
    dataset_id="cmems_obs-oc_glo_bgc-transp_my_l4-multi-4km_P1M",
    variables=["SPM", "ZSD"],
    minimum_latitude=min_lat, minimum_longitude=min_lon,
    maximum_latitude=max_lat, maximum_longitude=max_lon,
    start_datetime=start_date, end_datetime=end_date,
    file_format="zarr",
    output_filename = "cmems_obs-oc_glo_bgc-transp_my_l4-multi-4km_P1M.zarr",
    output_directory="./resources/copernicus_marine_service",
    chunk_size_limit=50_000_000,
    overwrite=True,
    dry_run=True # False to download and True to only simulate
)
```

```python
# Volume absorption coefficient of radiative flux in sea water due to dissolved organic matter and non algal particles CDM [m-1]
copernicusmarine.subset(
    dataset_id="cmems_obs-oc_glo_bgc-optics_my_l4-multi-4km_P1M",
    variables=["CDM"],
    minimum_latitude=min_lat, minimum_longitude=min_lon,
    maximum_latitude=max_lat, maximum_longitude=max_lon,
    start_datetime=start_date, end_datetime=end_date,
    file_format="zarr",
    output_filename = "cmems_obs-oc_glo_bgc-optics_my_l4-multi-4km_P1M.zarr",
    output_directory="./resources/copernicus_marine_service",
    chunk_size_limit=50_000_000,
    overwrite=True,
    dry_run=True # False to download and True to only simulate
)
```

```python
chl = xr.open_zarr("./resources/copernicus_marine_service/cmems_obs-oc_glo_bgc-plankton_my_l4-multi-4km_P1M.zarr", consolidated=True)
```

```python
# Target resolution (degrees)
res = 0.125

# Target lat/lon vectors for a resolution of 0.125 degrees
new_lons = np.arange(min_lon, max_lon + res, res)
new_lats = np.arange(min_lat, max_lat + res, res)

# Create a target grid as xarray Dataset
ds_tgt = xr.Dataset({
    'lat': (['lat'], new_lats),
    'lon': (['lon'], new_lons)})

# 'conservative' if you want area-weighted aggregation  #this needs to be explained in methods later
regridder = xe.Regridder(chl, ds_tgt, 'conservative', periodic=False)

# Apply regridding
chl_regrided = regridder(chl)

chl_regrided_mean = chl_regrided.mean(dim='time')

with ProgressBar():
    chl_regrided_mean = chl_regrided_mean.compute()

chl_regrided_mean["CHL"].plot()
```
